```python
import datetime
from dateutil.parser import parse

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="ticks")
sns.set_context(context="talk")
```

```python
data = pd.read_csv("../data/wellvue_o2_data/O2Ring-20200421205858-OXIRecord.csv")
data.head()
```

```python
data['Time'] = pd.to_datetime(data['Time'], format="%H:%M:%S %b %d，%Y") # Lol no clue what the weird comma is
data.head()
```

```python
ax = data.plot(x="Time", y="SpO2(%)", legend=False, linewidth=0.5, figsize=(15, 6))
ax2 = ax.twinx()
data.plot(x="Time", y="Pulse Rate(bmp)", ax=ax2, linewidth=0.5, legend=False, color="g")
ax.set(ylim=(80, 100))
plt.title("SpO2(%) and Pulse Rate(bmp) over one day")
ax.set_ylabel("SpO2(%)")
ax2.set_ylabel("Pulse Rate(bmp)", rotation=270, labelpad=20)
ax2.set(ylim=(40, 90))
ax.set_xlabel("Time")
```

```python
ax = data.plot(x="Time", y="SpO2(%)", legend=False, linewidth=0.5, figsize=(15, 6))
ax2 = ax.twinx()
data.plot(x="Time", y="Motion", ax=ax2, linewidth=0.5, legend=False, color="g")
ax.set(ylim=(80, 100))
plt.title("SpO2(%) and Motion over one day")
ax.set_ylabel("SpO2(%)")
ax2.set_ylabel("Motion", rotation=270, labelpad=20)
ax2.set(ylim=(0, 200))
ax.set_xlabel("Time")
```

```python
ax = data.plot(x="Time", y="Pulse Rate(bmp)", legend=False, linewidth=0.5, figsize=(15, 6))
ax2 = ax.twinx()
data.plot(x="Time", y="Motion", ax=ax2, linewidth=0.5, legend=False, color="g")
ax.set(ylim=(45, 90))
plt.title("Pulse Rate(bmp) and Motion over one day")
ax.set_ylabel("Pulse Rate(bmp)")
ax2.set_ylabel("Motion", rotation=270, labelpad=20)
ax2.set(ylim=(0, 300))
ax.set_xlabel("Time")
```
